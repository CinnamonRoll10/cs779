### **Background: Singular Value Decomposition (SVD)**

SVD is a matrix factorization technique in linear algebra. It decomposes a matrix M  into three matrices:  
$M = U \Sigma V^{*}$

Where:
- $M$ is the original matrix.
- $U$ is an $m \times m$  unitary matrix.
- $\Sigma$ is an $m \times n$  diagonal matrix with non-negative real numbers (singular values).
- $V^{*}$ is the conjugate transpose of $V$ , an $n \times n$ unitary matrix.

This decomposition is useful in reducing the dimensionality of data while retaining important features, making it a powerful tool for tasks like latent semantic analysis (LSA) (https://en.wikipedia.org/wiki/Latent_semantic_analysis) in Natural Language Processing (NLP). You can also watch this nice video lecture series on SVD: https://www.youtube.com/watch?v=gXbThCXjZFM

```python
import re
import pandas as pd

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.strip()

    return text

train_df['cleaned_text'] = train_df['text'].apply(clean_text)
test_df['cleaned_text'] = test_df['text'].apply(clean_text)

print(train_df[['text', 'cleaned_text']].head())
print(test_df[['text', 'cleaned_text']].head())
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
def work(text):
    doc = nlp(text)
    text2 = " ".join([token.lemma_ for token in doc if not token.is_stop])

    return text2

train_df_sampled['cleaned_text'] = train_df_sampled['cleaned_text'].apply(work)
test_df_sampled['cleaned_text'] = test_df_sampled['cleaned_text'].apply(work)

print(train_df_sampled[['text', 'cleaned_text']].head())
print(test_df_sampled[['text', 'cleaned_text']].head())
```

```python
from collections import defaultdict
from scipy.sparse import dok_matrix
import numpy as np

V = set()
for text in train_df_sampled['cleaned_text']:
    for word in text.split():
        V.add(word)
V = list(V)
n = len(V)

index = {}
for i in range(n):
  index[V[i]] = i

print(n)

# M = dok_matrix((n, n), dtype=np.int32)
# k = 5

# for text in train_df_sampled['cleaned_text']:
#   for i, word in enumerate(text.split()):
#     size = (k - 1) // 2
#     s = max(0, size)
#     e = min(len(text), i + size)
#     for j in range(s, e):
#       if i != j:
#         M[index[word], index[text.split()[j]]] += 1

# print(M.head())
```

```python
M = dok_matrix((n, n), dtype=np.int32)
k = 5
it = 0

for text in train_df_sampled['cleaned_text']:
  words = text.split()
  for i, word in enumerate(words):
    size = (k - 1) // 2
    s = max(0, i - size)
    e = min(len(words), i + size + 1)
    for j in range(s, e):
      if i != j:
        v1 = index[word]
        v2 = index[words[j]]
        it += 1
        if (it <= 10):
          print(v1, v2)
        M[v1, v2] += 1

print(M[5300, 1461])
```

```python
from sklearn.decomposition import TruncatedSVD
M_sparse = M.tocsr()

svd = TruncatedSVD(n_components=500)
U = svd.fit_transform(M_sparse)
Vt = svd.components_
sigma = svd.singular_values_

print("U shape:", U.shape)
print("Sigma shape:", sigma.shape)
print("Vt shape:", Vt.shape)
```

```python
import matplotlib.pyplot as plt

var = svd.explained_variance_ratio_
var_sum = np.cumsum(var)

plt.plot(var_sum)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance')
plt.title('Cumulative Variance vs. Number of Components')
plt.grid(True)
plt.show()
```

```python
threshold = 0.80
r = np.argmax(var_sum >= threshold)

print(r)
```

We chose r such that we keep atleast 80% variance in the data. Higher values of r leads to more precise results.

```python
svd = TruncatedSVD(n_components=r)
Ur = svd.fit_transform(M_sparse)
sigma_r = svd.singular_values_
Sr = np.diag(sigma_r)
Vtr = svd.components_

M2 = np.dot(Ur, Vtr)
print(f"M_ap shape : ", {M2.shape})
```

```python
word_embeddings = {}

print(len(index))
it = 0
for word, idx in index.items():
    word_vector = M[index[word], :].toarray().flatten()
    it += 1
    if (it % 1000 == 0):
      print(f"DONE {it}")

    embedding = np.linalg.inv(Sr) @ Ur.T @ word_vector
    word_embeddings[word] = embedding
```

```python
sample_word = "name"
print(word_embeddings.get(sample_word))
```

```python
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

sample_word_2 = ['happy', 'sad', 'mad', 'joy']

for i in range(4):
  for j in range(i + 1, 4):
    if word_embeddings.get(sample_word_2[i]) is not None and word_embeddings.get(sample_word_2[j]) is not None:
        print(f"Word1 {sample_word_2[i]}, Word2 {sample_word_2[j]}")
        print(cosine_similarity(word_embeddings[sample_word_2[i]], word_embeddings[sample_word_2[j]]))
```

**ANALYSIS**
- When a lower r (e.g., 50-60%) was chosen, the word embeddings captured only a limited amount of semantic information, resulting in less meaningful relationships. Words that were not semantically related appeared closer in the vector space.
A higher r (80-90%) led to more precise embeddings. The vector space more accurately reflected real-world semantic relationships between words, as seen in the higher cosine similarity values between words like happy and joy.
However, increasing r also significantly increased computational complexity and execution time. For very high r values, the results were more precise, but the cost of matrix operations and embedding calculations grew substantially.
- Optimal Threshold: Based on our experiments, an r value around 80% seemed to strike a good balance between computational efficiency and the quality of the embeddings.

- Effect of k (Neighbor Parameter in Similarity Analysis)
Experiment: We varied the neighbor parameter (k) used in cosine similarity analysis to identify the most similar words.
Results:
When k was low (e.g., 2-3), the embeddings tended to retrieve words that were very close in the vector space. This sometimes led to identifying words with slightly different meanings as being close, especially if they shared common contexts (e.g., happy and funny).
As k increased (e.g., around 9), the list of neighboring words for each target word became more representative of semantic relationships. The model began identifying synonyms and related terms more reliably.
Higher k values beyond 10 started introducing unrelated words into the neighborhood, suggesting that including too many neighbors dilutes the semantic specificity of the relationships.
- Optimal k: We found that a k value around 9 provided a good balance. It offered a sufficiently diverse set of semantically related words without introducing too many irrelevant terms.


Increasing r and k improved results but also came with higher computational costs.
