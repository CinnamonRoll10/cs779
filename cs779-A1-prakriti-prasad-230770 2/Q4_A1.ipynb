# **Question 4: N-Gram models**

Language models are statistical models that give out the probability distribution of a sentence (sequence of tokens). They rely on the joint probability distribution of the tokens in the sentence.

For example, let us calculate the probability of the sentence "Monza is a fast track on the F1 GP". Then the probability of the sentence is given as:
$$
P = P(\text{<s>Monza is a fast track on the F1 GP</s>})
$$
Where `<s>` and `</s>` are predefined start and end tokens respectively.

We can then divide the sentence into tokens and use the chain rule of probability to make the calculations easier.
$$
P = P(\text{Monza|<s>}).P(\text{is|<s>Monza}).P(\text{a|<s>Monza is}).P(\text{fast|<s>Monza is a}) ... .P(\text{</s>|<s>Monza is a fast track on the F1 GP})
$$

Learning the joint probability distirbution is much more difficult and hence usually we have Markov Assumption (https://en.wikipedia.org/wiki/Discrete-time_Markov_chain). The main idea is the current token only depend on the previous token and you can ignore the history before the previous token. Given the Markov Assumption, we have:

$$
P(\text{is|<s>Monza}) = P(\text{is|Monza}) \\
P(\text{a|<s>Monza is}) = P(\text{a|is}) \\
P(\text{fast|<s>Monza is a}) = P(\text{fast|a}) \\
... \\
P(\text{</s>|<s>Monza is a fast track on the F1 GP}) = P(\text{</s>|GP})
$$

Hence, we can write the probability of the complete sentence as

$$
P(\text{<s>Monza is a fast track on the F1 GP</s>}) = \\P(\text{Monza|<s>}).P(\text{is|Monza}).P(\text{a|is})...P(\text{</s>|GP})
$$

Above we made 1st order Markov Assumption, giving us a **BiGram Language Model**. Similarly, we can make $(n-1)^{st}$ order Markov assumption, giving us  **n-gram Language Model**

In general using n-gram model, the probability of a sentence $s$ is given as:
$$
P(s) = \prod{P(w_n|w_{1} w_{2} w_{3} w_{4} ... w_{n-1})}
$$

```python
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm", disable=["ner", "parser", "lemmatizer"])

def get_pos_tags(text):
    doc = nlp(text)
    return [token.pos_ for token in doc]

df1['pos_tags'] = df1['text'].apply(get_pos_tags)
print(df1[['text', 'pos_tags']].head())
```

```python
def append_start_end(pos_tags):
    return ['START'] + pos_tags + ['END']

df1['pos_tags'] = df1['pos_tags'].apply(append_start_end)
print(df1[['text', 'pos_tags']].head())
```

```python
from collections import defaultdict

cnt1 = defaultdict(int)
cnt2 = defaultdict(int)
cnt3 = defaultdict(int)
total = 0

for x in df1['pos_tags']:
  for i in range(len(x)):
    cnt1[x[i]] += 1
    total += 1
  for i in range(len(x) - 1):
    bigram = (x[i], x[i + 1])
    cnt2[bigram] += 1
  for i in range(len(x) - 2):
    trigram = (x[i], x[i + 1], x[i + 2])
    cnt3[trigram] += 1

print(total)
```

```python
prob1 = defaultdict(float)
prob2 = defaultdict(float)
prob3 = defaultdict(float)

for w1, cnt in cnt1.items():
  prob1[w1] = cnt / total

for (w1, w2), cnt in cnt2.items():
  prob2[(w1, w2)] = cnt / cnt1[w1]

for (w1, w2, w3), cnt in cnt3.items():
  prob3[(w1, w2, w3)] = cnt / cnt2[(w1, w2)]

print(prob1.items())
print(prob2.items())
print(prob3.items())
```

**Proof of Formula**

Let's recall the definition of a bigram probability. Say the bigram is $(x, y)$. Then it is the probability $w_{i + 1} = y$ given that $w_i = x$.

This follows directly from the definition of conditional probability. The total number of ways is the number of times $w_i = x$ and $w_{i + 1}$ exists. The number of good ways for our bigram is the number of times $w_i = x$ and $w_{i + 1} = y$.

Thus, the probability can be calculated as $\frac{C(x, y)}{C(x, *)}$ where $C(x, y)$ represents the count of times $(x, y)$ appears back to back, and $*$ represents a wildcard character (any character will satisfy).

Similarly, for a trigram $(x, y, z)$, the probabiliy can be written as $\frac{C(x, y, z)}{C(x, y, *)}$

```python
# Backup Model

lambda_val = 0.4

def get_uni_prob(w1):
  return prob1[w1]

def get_bi_prob(w1, w2):
  v = prob2[(w1, w2)]
  if v == 0:
    v = lambda_val * get_uni_prob(w2)
  return v

def get_tri_prob(w1, w2, w3):
  v = prob3[(w1, w2, w3)]
  if v == 0:
    v = lambda_val * get_bi_prob(w2, w3)
  return v

list_items = []
for x, p in prob1.items():
  list_items.append(x)

arr_fixed_probs = []
for i in range(len(list_items)):
  for j in range(len(list_items)):
    for k in range(len(list_items)):
      v = get_tri_prob(list_items[i], list_items[j], list_items[k])
      v2 = prob3[(list_items[i], list_items[j], list_items[k])]
      if v != v2:
        arr_fixed_probs.append((i, j, k, v, v2))

print(f"Wrong Triplets", len(arr_fixed_probs))
for (i, j, k, v1, v2) in arr_fixed_probs[:5]:  # Iterate over the first 5 elements
    print(f"Example of Fixed Probability of Trigram ({list_items[i]}, {list_items[j]}, {list_items[k]})")
    print(f"New Assigned Probability: {v1}, old was {v2}")
```

Implemented a Backup Model and Smoothing Approach by reading the paper linked. They both try to fix the 0 probability events by different methods, so that a sentence with a never seen POS tag sequence doesn't give 0 probability.

Backup Model does it by falling back to Bigram Model when Trigram doesn't work and Unigram when Bigram doesnt work. Smoothing adds a small k to every element, we chose k = 0.1 here.

```python
# Smoothing
import copy
k_to_add = 0.1

smooth_prob1 = copy.deepcopy(prob1)
smooth_prob2 = copy.deepcopy(prob2)
smooth_prob3 = copy.deepcopy(prob3)

total = 1

for x in list_items:
  smooth_prob1[x] += k_to_add
  total += smooth_prob1[x]

for x in list_items:
  smooth_prob1[x] /= total

for x in list_items:
  total = 0
  for y in list_items:
    smooth_prob2[(x, y)] += k_to_add
    total += smooth_prob2[(x, y)]
  for y in list_items:
    smooth_prob2[(x, y)] /= total

for x in list_items:
  for y in list_items:
    total = 0
    for z in list_items:
      smooth_prob3[(x, y, z)] += k_to_add
      total += smooth_prob3[(x, y, z)]
    for z in list_items:
      smooth_prob3[(x, y, z)] /= total

sentence_example_2 = "My name is Prakriti Prasad"
print(f"Example Sentence is {sentence_example_2}")

arr_pos_2 = get_pos_tags(sentence_example_2)
arr_pos_2 = ['START'] + arr_pos_2 + ['END']
print(arr_pos_2)

prob_tri = 1
smooth_prob_tri = 1
backup_prob_tri = 1

for i in range(len(arr_pos_2) - 2):
  prob_tri *= prob3[(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])]
  backup_prob_tri *= get_tri_prob(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])
  smooth_prob_tri *= smooth_prob3[(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])]

print(f"Normal Trigram Probability is {prob_tri}")
print(f"Smoothed Trigram Probability is {smooth_prob_tri}")
print(f"Backup Trigram Probability is {backup_prob_tri}")

print()

sentence_example_2 = "Oh, she Wow" # Example of a sentence with initial 0 probability
print(f"Example Sentence is {sentence_example_2}")

arr_pos_2 = get_pos_tags(sentence_example_2)
arr_pos_2 = ['START'] + arr_pos_2 + ['END']
print(arr_pos_2)

prob_tri = 1
smooth_prob_tri = 1
backup_prob_tri = 1

for i in range(len(arr_pos_2) - 2):
  prob_tri *= prob3[(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])]
  backup_prob_tri *= get_tri_prob(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])
  smooth_prob_tri *= smooth_prob3[(arr_pos_2[i], arr_pos_2[i + 1], arr_pos_2[i + 2])]

print(f"Normal Trigram Probability is {prob_tri}")
print(f"Smoothed Trigram Probability is {smooth_prob_tri}")
print(f"Backup Trigram Probability is {backup_prob_tri}")
```
