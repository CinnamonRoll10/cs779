## Background: What is Zipf's Law?

Zipf’s Law states that in a large corpus of natural language, the frequency of any word is inversely proportional to its rank in the frequency table. This implies that frequency and rank are related by a power law.


Zipf's Law can be expressed as:
$$
f(r) = \frac{C}{r^k}
$$
Where:
- \( f(r) \) is the frequency of the word at rank \( r \).
- \( r \) is the rank of the word (1 for the most frequent word, 2 for the second most frequent, etc.).
- \( C \) is a constant that depends on the distribution of the frequencies.
- \( k \) is a parameter that decides the power law relationship.

For example, for some given constant 'k'

- The most frequent word will appear roughly twice as often as the second most frequent word.
- The second most frequent word will appear approximately three times as often as the third most frequent word, and so on.

```python
def calculate_word_frequencies(tokens):
    word_freq = {}
    for i in range(len(tokens)):
        token = tokens[i]
        if token in word_freq:
            word_freq[token] += 1
        else:
            word_freq[token] = 1
    return word_freq
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_word_frequencies(sorted_word_freq):
    ranked_words = sorted_word_freq
    ranks = np.arange(1, len(ranked_words) + 1)

    frequency_list = []
    for word, freq in ranked_words:
      frequency_list.append(freq)

    frequencies = np.array(frequency_list)
    plt.plot(ranks, frequencies, label='Actual Data')
    k = 1
    scaling_factor = 1 / (ranks ** k)
    theoretical_zipf = scaling_factor * frequencies[0]

    plt.plot(ranks, theoretical_zipf, label=f'Zipf\'s Law (k={k})', linestyle='--')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.title('Rank vs Frequency (Actual vs Zipf\'s Law)')
    plt.legend()
    plt.show()

    # Log-log graph
    log_ranks = np.log(ranks)
    log_frequencies = np.log(frequencies)

    coefficients = np.polyfit(log_ranks, log_frequencies, 1)
    slope = coefficients[0]
    intercept = coefficients[1]

    print("k:", slope)

    plt.plot(log_ranks, log_frequencies, label="Log Rank vs Log Frequency")
    plt.plot(log_ranks, slope * log_ranks + intercept, color='red')
    plt.xlabel('Log Rank')
    plt.ylabel('Log Frequency')
    plt.title('Log Rank vs Log Frequency with Regression Line')
    plt.legend()
    plt.show()

```

```python
import collections
from indicnlp.tokenize import indic_tokenize

def chunkify(file_path, chunk_size):
    with open(file_path, 'r', encoding='utf-8') as file:
        chunks = [chunk for chunk in iter(lambda: file.read(chunk_size), '')]
    return chunks

def update_word_frequencies(main_freq, new_freq):
    for word, count in new_freq.items():
        main_freq[word] += count

def process_chunk(chunk, lang='bn'):
    tokens = indic_tokenize.trivial_tokenize(chunk, lang=lang)
    word_freq = collections.Counter(tokens)
    return word_freq

def calculate_total_word_frequencies(file_path, chunk_size=10000):
    total_freq = collections.Counter()
    chunks = chunkify(file_path, chunk_size)

    for chunk in chunks:
        chunk_freq = process_chunk(chunk)
        update_word_frequencies(total_freq, chunk_freq)

    return total_freq

word_frequencies_ta = calculate_total_word_frequencies(file_path_bn, chunk_size=10000)
print(word_frequencies_bn.most_common(10))
'''The code is designed to process a large text file by breaking it into manageable
chunks to analyze word frequencies efficiently. The chunkify function reads the file
in pieces of a specified size and collects these pieces into a list. Then, calculate_total_word_frequencies
 processes each chunk by tokenizing it into words and counting their occurrences. It updates a cumulative count
 of words across all chunks using the update_word_frequencies function. Finally, the function returns the total
 word frequency distribution. This approach helps handle large files by not loading the entire file into memory
 at once.
'''
```

The core idea behind chunking is to split the large file into smaller, manageable pieces or "chunks," each of which can be processed independently. In this case, the chunkify function reads the file in chunks of a specified size (e.g., 10,000 characters). Instead of loading the entire file at once, only a small portion is loaded into memory, processed for tokenization and word frequency counting, and then discarded before the next chunk is processed. This reduces the overall memory footprint significantly.

Moreover, tokenizing and calculating word frequencies in smaller chunks does not compromise accuracy. Each chunk is tokenized using the Indic NLP library's trivial_tokenize, which allows for language-specific tokenization. The word frequencies from each chunk are then combined into a running total by updating the total_freq counter using the update_word_frequencies function. This incremental approach ensures that, despite processing the corpus in smaller parts, the final result represents the complete word frequency distribution for the entire dataset.

## What are Regular Expressions?
Regular expressions (regex or regexp) (https://en.wikipedia.org/wiki/Regular_expression) are powerful tools for finding or matching patterns in strings. They allow you to define specific patterns and search for characters or words within text.


## Regular expressions in Python
Python has a built-in package called `re` (https://docs.python.org/3/library/re.html), which can be used to work with Regular Expressions. Using this module, you specify the rules for the set of possible strings that you want to match; this set might contain English sentences, or e-mail addresses, or TeX commands, or anything you like. You can then ask questions such as “Does this string match the pattern?”, or “Is there a match for the pattern anywhere in this string?”. You can also use REs to modify a string or to split it apart in various ways.

## Rules

Regular expressions can contain both special and ordinary characters. Most ordinary characters, like `'A'`, `'a'`, or `'0'`, are the simplest regular expressions; they simply match themselves. You can concatenate ordinary characters, so `last` matches the string `'last'`.

Some characters, like `'|'` or `'('`, are special. Special characters either stand for classes of ordinary characters, or affect how the regular expressions around them are interpreted.

```python
import re
import pandas as pd
import os


def find_words(text):
    return re.findall(word_pattern, text)

def process_chunk(chunk):
    all_words = []
    for message in chunk['message']:
        words = find_words(message)
        all_words.extend(words)
    return all_words

def process_in_chunks(dataframe, chunk_size=1000, output_file='words.txt'):
    num_chunks = len(dataframe) // chunk_size + 1

    with open(output_file, 'w', encoding='utf-8') as file:
        for i in range(num_chunks):
            start = i * chunk_size
            end = start + chunk_size
            chunk = dataframe[start:end]
            chunk_words = process_chunk(chunk)

            file.writelines(f"{word}\n" for word in chunk_words)

            print(f"Processed chunk {i + 1} of {num_chunks}")

process_in_chunks(dataframe, chunk_size=1000, output_file='words.txt')
```

```python
import collections

def process_chunk(chunk, word_freq_counter):
    words = chunk.splitlines()
    word_counter = collections.Counter(words)
    word_freq_counter.update(word_counter)

def process_file_in_chunks(file_path, chunk_size=1024 * 1024):  # Default chunk size 1MB
    word_freq_counter = collections.Counter()

    with open(file_path, 'r', encoding='utf-8') as file:
        while True:
            chunk = file.read(chunk_size)
            if not chunk:
                break
            process_chunk(chunk, word_freq_counter)
            print("Processed a chunk")

    return word_freq_counter

file_path = 'words.txt'
word_frequencies = process_file_in_chunks(file_path, chunk_size=1024 * 1024)  # Process in 1MB chunks
print(f"Total number of unique words: {len(word_frequencies)}")
print(word_frequencies.most_common(10))
```
